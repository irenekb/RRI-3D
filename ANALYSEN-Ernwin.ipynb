```python
colorsv=['cornflowerblue','seagreen','orange']
colorse=['cornflowerblue','green','orange','red']

def countingfunction1 (count1, count2):
    if count2 == 1:
        count1 += 1
        count2=0
    else:
        count2 += 1
    
    return count1, count2

def countingfunction2(count2):
    if count2 == 1:
        count2 = 0
    else:
        count2 = 1
        
    return count2

def numbers(nr1, nr2, switch):
    if nr2 == 1:
        nr1 += 1
    nr2 = countingfunction2(nr2)
    switch = countingfunction2(switch)
    
    return nr1, nr2, switch
```
